````python
# Get access to a specific pixel

(blue, green, red) = img[100, 100]
print(f"blue={blue}, green={green}, red={red}")
````

#### Cut a part of image

````python
cut_img = img[50:350, 50:350]
cv2_imshow(cut_img)
````

#### Resize image

````python
resized_cut_img = cv2.resize(cut_img, (500, 500))
cv2_imshow(resized_cut_img)
````

Case when we don't have height and width.

````python
test2 = img[50:350, 50:450]
cv2_imshow(test2)
````

````python
height = 150

h, w = test2.shape[0:2]

ratio = w / h
width = int(height * ratio)

resized_test2 = cv2.resize(test2, (width, height))
cv2_imshow(resized_test2)
````

````python
import imutils

resized = imutils.resize(test2, height=height)
cv2_imshow(resized)
````

````python
rotated = imutils.rotate(resized, -45)
cv2_imshow(rotated)
````

#### Blur image

````python
blurred = cv2.GaussianBlur(resized, (11, 11), 0)
cv2_imshow(blurred)
````

````python
import numpy as np

summing = np.hstack((resized, blurred))
cv2_imshow(summing)
````

#### Drawings

````python
cv2.rectangle(resized, (100, 100), (250, 150), (0, 0, 255), 2)
cv2_imshow(resized)
````

````python
cv2.line(resized, (0, 0), (200, 200), (255, 0, 0), 5)
cv2_imshow(resized)
````

````python
black_img = np.zeros((200, 200, 3), np.uint8)

points = np.array([[0, 0], [100, 50], [50, 100], [0, 0]])
cv2.polylines(black_img, np.int32([points]), 1, (255, 255, 255))
cv2_imshow(black_img)
````

````python
black_img = np.zeros((200, 200, 3), np.uint8)

cv2.circle(black_img, (100, 100), 50, (0, 0, 255), 2)
cv2_imshow(black_img)
````

````python
black_img = np.zeros((200, 550, 3), np.uint8)

font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.putText(
    black_img, 'OpenCV', (0, 100), font, 4, (255, 255, 255), 4, cv2.LINE_4
)
cv2_imshow(black_img)
````
